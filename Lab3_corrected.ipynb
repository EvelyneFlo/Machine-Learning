<h3>Podziel zbiór na treningowy i testowy używając train_test_split.
Jako zmienną niezależną ustaw dane składające się z kolumn o nazwach

<h3>Wykonaj kodowanie zmiennych kategorycznych do zmiennych
liczbowych. Wykorzystaj pętlę for i metodę enumerate(). Wynik powinien
być podobny do tego na obrazku poniżej (UWAGA! Na rysunku jest
tylko część rozwiązania i przykład tylko dla zmiennej kabina).<h3>

```python
cabin_dict={}
for i, value in enumerate(df.cabin.unique()):
    cabin_dict[value]=i
cabin_dict
```

```python
cabinReduced_dict={}
for i, value in enumerate(df.cabinReduced.unique()):
    cabinReduced_dict[value]=i
cabinReduced_dict
```

```python
sex_dict={}
for i, value in enumerate(df.sex.unique()):
    sex_dict[value]=i
sex_dict
```

<h3>Zastąp etykiety zmiennej (tu przykład dla kabina) słownikiem
stworzonym w kroku 6. Do tego będzie potrzebne mapowanie.
<h3>

```python
X_train["cabin_map"] = X_train["cabin"].map(cabin_dict)
X_train[["cabin", "cabin_map"]].head(20)
```

```python
X_test["cabin_map"] = X_test["cabin"].map(cabin_dict)
X_test[["cabin", "cabin_map"]].head(20)
```

```python
X_train["cabinReduced_map"] = X_train["cabinReduced"].map(cabinReduced_dict)
X_train[["cabinReduced", "cabinReduced_map"]].head(20)
```

```python
X_test["cabinReduced_map"] = X_test["cabinReduced"].map(cabinReduced_dict)
X_test[["cabinReduced", "cabinReduced_map"]].head(20)
```

```python
X_train["sex_map"] = X_train["sex"].map(sex_dict)
X_train[["sex", "sex_map"]].head(20)
```

```python
X_test["sex_map"] = X_test["sex"].map(sex_dict)
X_test[["sex", "sex_map"]].head(20)
```

<h3>Zastąp brakujące wartości liczbą 0. Czy jest to najlepsze wyjście?

```python
cabin_dict.update({np.nan:0, 'B5':6})
cabin_dict
```

```python
cabinReduced_dict.update({'n':0, 'B':5})
cabinReduced_dict
```

```python
X_test["cabin_map"] = X_test["cabin"].map(cabin_dict)
X_train["cabin_map"] = X_train["cabin"].map(cabin_dict)
X_test["cabinReduced_map"] = X_test["cabinReduced"].map(cabinReduced_dict)
X_train["cabinReduced_map"] = X_train["cabinReduced"].map(cabinReduced_dict)
X_train.head(50)
```

<h6>
    
    Zastąpienie braków wartości liczbą 0 będzie dobrym wyjściem, gdy ta liczba nie występowała wcześniej w danym zbiorze danych. W innym wypadku możemy nadać błędne wartości zmiennym i tym samym zniekształcić model czy statystyki.
    W tym wypadku liczba 0 została już przypisana innej wartości, stąd samo zastąpienie braków danych liczbą 0 nie jest dobrym pomysłem - należy zamienić wartości dla NaN oraz klucza dla którego wartość jest równa 0
